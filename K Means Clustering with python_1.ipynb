```python
#Jakkamma
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#artificially creating the dataset
from sklearn.datasets import make_blobs
#make_blobs --> groups of data.
```

```python
#feature --> no of different columns
#random_state --> Determines random number generation for dataset creation. Pass an int
   #for reproducible output across multiple function calls
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8,
                 random_state=101)
```

```python
# when printing data we will get TUPLE.. So, we are grabbing first item in the data and output will be a numpy array
#tuple(x,y)
#here x --> datas, y-->cluster it belongs to
data[0]
```

```python
data[0].shape
```

```python
sns.set()
plt.scatter(data[0][:,0], data[0][:,1])
```

```python
data[1] #Have vales of clustyer that belongs to
```

```python
# c -->color, sequence, or sequence of color, optional
#plt.grid()

plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

```python
#Creating K-Means Clustering algoritms
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
#lables --> believes to be true for the clusters.
kmeans.labels_
```

```python
#Comparing our prdictions with original
#data[1] --> original cluster ..... we have created through make_blobs
#kmeans.labels --> predicted through kmeans algorithm
#*************but real time you wouldnt find original cluster
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title('original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data[0])
kmeans.cluster_centers_
kmeans.labels_
```
